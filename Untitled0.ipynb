```python
import pandas as pd
movies = pd.read_csv('real.csv')
```

```python
movies = movies[['id', 'original_title', 'overview', 'genres']]
```

```python
movies
```

```python
movies.loc[:,'tags'] = movies['overview'] + movies['genres']
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview', 'genres'])
```

```python
new_data
```

```python
!pip install --upgrade scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_cosine = cosine_similarity(vector)
```

```python
similarity_cosine
```

```python
new_data[new_data['original_title'] == 'Insurgent'].index[0]
```

```python
distance = sorted(list(enumerate(similarity_cosine[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].original_title)
```

```python
def recommend(movies):
  index = new_data[new_data['original_title'] == movies].index[0]
  distance = sorted(list(enumerate(similarity_cosine[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].original_title)
```

```python
recommend ("Iron Man")
```
